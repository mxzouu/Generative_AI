## 2. Check that all required packages import

These are the packages listed in `requirements.txt`. If any import fails, install it with `pip install -r requirements.txt` (in your activated environment).

````python
import importlib

# Map: friendly name -> module to import
packages = {
    "pandas": "pandas",
    "numpy": "numpy",
    "anthropic": "anthropic",
    "python-dotenv": "dotenv",
    "sentence-transformers": "sentence_transformers",
    "chromadb": "chromadb",
    "scikit-learn": "sklearn",
    "matplotlib": "matplotlib",
    "mcp": "mcp",
    "jupyter": "jupyter",
}

all_ok = True
for name, module in packages.items():
    try:
        mod = importlib.import_module(module)
        version = getattr(mod, "__version__", "(no version attribute)")
        print(f"✅ {name:<22} {version}")
    except ImportError as e:
        all_ok = False
        print(f"❌ {name:<22} NOT installed — {e}")

print()
print("✅ All packages imported successfully." if all_ok else "⚠️  Some packages are missing — run: pip install -r requirements.txt")
````

## 3. Test the Anthropic API (Claude Haiku)

This is the channel **your code** uses at runtime. It is billed per token on a **limited budget**, so we use **Claude Haiku** (the cheapest model) for everything in this course.

We keep the API key in a `.env` file at the **project root** and load it with `python-dotenv`. Your `.env` should contain a single line:

```
ANTHROPIC_API_KEY=sk-ant-your-key-here
```

`load_dotenv()` reads that file into the environment, and the SDK then picks up `ANTHROPIC_API_KEY` automatically.

> **Never commit your API key.** The `.env` file is already ignored by `.gitignore`.

````python
from dotenv import load_dotenv
import anthropic

# Load variables from the .env file at the project root into the environment.
# load_dotenv() searches up the directory tree, so it works from this notebooks/ subfolder.
load_dotenv()

# The SDK then reads ANTHROPIC_API_KEY from the environment automatically.
client = anthropic.Anthropic()

response = client.messages.create(
    model="claude-haiku-4-5",  # cheapest model — use this in your code
    max_tokens=100,
    messages=[
        {"role": "user", "content": "In one short sentence, confirm my setup works and wish me luck for the course."}
    ],
)

print(response.content[0].text)
print("\n✅ Anthropic API call succeeded.")
````

## 4. Test the local embedding model

Anthropic does not provide an embedding API, so for embeddings we use an open-source model that runs **locally and for free** via `sentence-transformers`.

The first run **downloads the model** (a few seconds to a minute, depending on your connection). Later runs are instant.

````python
from sentence_transformers import SentenceTransformer

# Small, fast model \u2014 good enough for the labs.
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

embedding = embed_model.encode("Hello, Generative AI!")

print(f"Embedding dimension: {len(embedding)}")
print(f"First 5 values: {embedding[:5]}")
print("\n\u2705 Embedding model loaded and working.")
````
